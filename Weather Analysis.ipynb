## Data Visualization

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Rel Hum_%', bins=20, kde=True,color='purple')
plt.title('Distribution of Humidity')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Temp_C', bins=20, kde=True,color='purple')
plt.title('Distribution of Temperature (Celsius)')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
wind_speed = df['Wind Speed_km/h'].unique()
print('Wind Speed km/h:-',wind_speed)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temp_C', y='Dew Point Temp_C', data=df, color='purple')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Dew Point Temperature (Â°C)')
plt.title('Temperature vs. Dew Point Temperature')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df,x='Wind Speed_km/h', color='purple')
plt.xlabel('Wind Speed (km/h)')
plt.title('Wind Speed Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Press_kPa', bins=20, kde=True,color='purple')
plt.title('Distribution of Pressure')
plt.xlabel('Pressure')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Visibility_km', bins=20, kde=True,color='purple')
plt.title('Distribution of Visibility')
plt.xlabel('Visibility')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df['Year'] = df['Date/Time'].dt.year
df['Month'] = df['Date/Time'].dt.month
df['Day'] = df['Date/Time'].dt.day
df['Hour'] = df['Date/Time'].dt.hour
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='Month', y='Temp_C', data=df, color='purple')
plt.title('Average Temperature Over Months')
plt.xlabel('Month')
plt.ylabel('Average Temperature (Â°C)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date/Time', y='Wind Speed_km/h', data=df,color='purple')
plt.title('Wind Speed Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Wind Speed (km/h)')
plt.xticks(rotation=45)
plt.show()
```
